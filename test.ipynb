```python
def countPrimes(n):
        """
        :type n: int
        :rtype: int
        """
        
        prime_count = 1 # this is 2
        
        if n<3:
            return 0
        
        else:
            n = tuple(filter(lambda x: (x%2!=0), range(2,n)))
            temp = n[:]
            # do not need to scan whole list, because n[i] = p x q = q x p for p < q, 
            # even if p = q, we only need to scan up to p = sqrt(n[i])
            n = tuple(filter(lambda x: x<=int((n[-1])**0.5), n))
            for i in range(len(n)):
                before = len(temp)
                temp = tuple(filter(lambda x: x%n[i]!=0, temp))
                if len(temp) < before:
                    prime_count+=1

            # rest of the numbers in temp are also prime!
            return prime_count + len(temp)
```

```python
%%time
countPrimes(100000)
```

```python
def get_primes(n):
    numbers = set(range(n, 1, -1))
    print(numbers)
    primes = []
    while numbers:
        p = numbers.pop()
        print(p)
        primes.append(p)
        print(set(range(p*2, n+1, p)))
        numbers.difference_update((range(p*2, n+1, p)))
    return len(primes)
```

```python
def generalizedGCD(num, arr):
    # WRITE YOUR CODE HERE
    # pass
    gcd = min(arr)
    
    if sum(map(lambda x: x%gcd, arr))==0:
        return gcd
    else:
        gcd = gcd//2
        while gcd>1:
            # print(gcd)
            remainder_sum = sum(map(lambda x: x%gcd, arr))
            if remainder_sum==0:
                return gcd
            else:
                gcd-=1
    return gcd
```

```python
def str_or_int(name):
    try:
        int(name)
        return int
    except ValueError:
        return str

def prioritizedOrders(numOrders, orderList):
    # WRITE YOUR CODE HERE
    # pass
    order_swap = list(map(lambda x: x.split(' '), orderList))
    prime_order = list(filter(lambda x: str_or_int(x[1])==str, order_swap))
    nonprime_order = list(filter(lambda x: str_or_int(x[1])==int, order_swap))
    prime_order.sort(key=lambda x: (x[1:], x[0]))
#     prime_order_check = prime_order[1:]
    
    new_prime = list(map(lambda x: ' '.join(x), prime_order))
    new_nonprime = list(map(lambda x: ' '.join(x), nonprime_order))
    
    new_order_list = new_prime + new_nonprime
    
    return new_order_list
```

```python
def isPalindrome(s):
        """
        :type s: str
        :rtype: bool
        """
        
        if s=='':
            return True
        else:
            temp = ''.join(list(filter(lambda char: char not in " ,_:;.!?'-", s)))
            temp = [str(x.lower()) for x in temp]
            l = len(temp)
            if l%2==0:
                print(int(l/2))
                print(temp[:int(l/2)])
                print(temp[:int(l/2)-1:-1])
                return temp[:int(l/2)]==temp[:int(l/2)-1:-1]
            elif l%2==1:
#                 print(l)
                print(temp)
                print(temp[:int(l/2)])
                print(temp[:int(l/2):-1])
                return temp[:int(l/2)]==temp[:int(l/2):-1]
```
